```python
%matplotlib inline
import numpy as np
import pandas as pd
import scipy as sp
import math
import matplotlib.pyplot as plt
from scipy import signal
from scipy import interpolate
import os

#header_IK = 8
```

### Generate Inverse Kinematics file

```python

# Read the CSV file
data = pd.read_csv('IMU-IK-ID_200hz.csv')

# Extract the 'Time' column
time_IK = data['Time']
time_seq_IK = np.array(time_IK)

# Extract the column headers
Label_IK_values = data.columns.tolist()

# Create a new DataFrame with 'Time' column
converted_IKdata = pd.DataFrame({'Time': time_IK})

# Generate the new time sequence
time_new_IK = np.arange(time_seq_IK[0], time_seq_IK[-1], 0.05)

# Loop over the remaining columns
for column in Label_IK_values[1:]:
    # Extract the column data
    IKdata_out = data[column].values
    
    # Interpolate the data
    interpolated_IKdataout = interpolate.interp1d(time_seq_IK, IKdata_out, kind='cubic')
    IKdataout_new = interpolated_IKdataout(time_new_IK)
    
    # Create a new DataFrame with interpolated data
    IKdataout_i = pd.DataFrame({column: IKdataout_new})
    
    # Concatenate the new DataFrame with the converted data
    converted_IKdata = pd.concat([converted_IKdata, IKdataout_i], axis=1)

# Compute summary statistics
converted_IKdata.describe()
```

```python
IK_output_name = 'check.csv'
converted_IKdata.to_csv(IK_output_name, index = False)
```
